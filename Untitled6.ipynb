```python
df.info()
```

```python
df['waterfront'].value_counts()
```

```python
data_1 = df['waterfront'].value_counts()
names = data_1.index
values = data_1.values
```

```python
plt.figure(figsize=(8, 6))
plt.pie(values, autopct='%1.1f%%', labels=names)
plt.title('Распределение домов по виду на набережную')
plt.legend(names, loc='upper right');
```

###Вывод
За обозначение дома без вида на набережную берем 0.
За обозначение дома с видом на набережную берем 1.
(Другого не обозначено в таблице)
